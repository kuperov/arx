# Supplementary results table


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
supp = pd.read_csv('../results/supplementary.csv')
alphas = sorted(supp.alpha.unique())
```

```python
supp.head()
```

```python
supp.scheme.unique()
```

```python
schemes = ['LOO', 'hv-block (h=3, v=3)']
```

```python
table = []
for i in range(1, 6+1):
    for s in schemes:
        rec = {
            'experiment': f'Experiment {i}',
            'scheme': s,
        }
        for v in ['easy', 'hard']:
            for a in alphas:
                err = supp[(supp.ex_no == i) & (supp.alpha == a) & (supp.scheme == s) & (supp.variant == v)].negshare
                rec[f'{v} {a:.2f}'] = 100*round(float(err),3) if len(err) else None
        table.append(rec)

tabledf = pd.DataFrame(table)
tabledf
```
